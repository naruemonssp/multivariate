```
def function_name(_Input_) :
  do_something with _Input_ to get _Output_
  return _Output_
```

Function มีส่วนสำคัญทั้งหมด 4 ส่วน


1.   บอก python ว่าเราจะเขียนฟังก์ชั่น ชื่ออะไร def function_name(): (ขาดไม่ได้)
2.   กำหนดตัวแปรที่จะเป็น input *Input*(ขาดได้)
3. ส่วนประมวลผล do_something with Input to get Output (ขาดไม่ได้)
4. ส่วน output return *Output*(ขาดได้)

เราใช้ เว้นวรรค(indent) เพื่อ  


```
def print_name(name,surname,ID):
  st=f'ชื่อ{name} นามสกุล {surname} ID {ID}'
  return st
print('outside function')
```

การเรียกใช้ฟังก์ชั่น

```
print_name('นฤมล','ไสยโสภณ','623021050-6')
```

```
output_string = print_name('นฤมล','ไสยโสภณ','623021050-6')
print(output_string)
```

ฟังก์ชั่นไม่จำเป็นต้องมี output

```
def print_name2(surname,ID,name):
  st = f'ชื่อ {name} นามสกุล {surname} รหัส {ID}'
  print(st)
```

```
print_name2('นฤมล','ไสยโสภณ','623021050-6')
```

```
op2 = print_name2('นฤมล','ไสยโสภณ','623021050-6')
```

```
print(op2)
```

## 1. Input ของ function ใน python มีสองแบบ input ที่จำเป็นต้องใส่กับ input ที่ไม่จำเป็นต้องใส่ (มีค่า default)
## 2. เราต้องเรียง input ที่จำเป็นต้องใส่ขึ้นก่อน

```
def print_2line_default(name,surname,ID,grade='F'):
  st = f'ชื่อ {name} นามสกุล {surname} รหัส {ID}'
  print(st)
  st2 = f'เกรดวิชา Multivariate >>> {grade}'
  print(st2)
```

```
print_2line_default('นฤมล','ไสยโสภณ','623021050-6','A')
```

เราสามารถเรียกฟังก์ชั่นโดยใช้ชื่อตัวแปรของฟังก์ชั่นช่วยได้

```
print_name2(name='นฤมล', surname='ไสยโสภณ', ID='623021050-6')
```
